```python
import matplotlib.pyplot as plt

# Dados de exemplo (tipos de alimentos e quantidades)
sentimentos = ['alegria', 'tristeza', 'medo', 'raiva', 'nojo']
expressao = [29.19, 54.53, 49.20, 28.61, 45.40]

# Plotar o gráfico de pizza
plt.pie(expressao, labels=sentimentos, autopct='%1.1f%%', startangle=140)

# Personalização do gráfico
plt.title('Emoções ligadas ao Tornado em Nebraska, cobertura pela CNN')

# Mostrar o gráfico
plt.savefig ("grafiv.png")
```
